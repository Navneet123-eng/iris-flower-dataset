# Feature Engineering for Iris Dataset

## 1. Introduction
In this notebook, we will perform feature engineering on the Iris dataset. The goal is to improve the performance of our machine learning model by creating new features, transforming existing ones, and selecting the most relevant features.

## 2. Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.decomposition import PCA
from sklearn.feature_selection import SelectKBest, chi2, RFE
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from mpl_toolkits.mplot3d import Axes3D
```

## 3. Load the Dataset

```python
# Load the Iris dataset
from sklearn.datasets import load_iris

iris = load_iris()
data = pd.DataFrame(data=iris.data, columns=iris.feature_names)
data['target'] = iris.target

# Display the first few rows of the dataset
data.head()
```

# 4. Data Preprocessing
## 4.1 Handling Missing Values

```python
# Check for missing values
data.isnull().sum()

# Iris dataset does not have missing values, so no further action is required
```

## 4.2 Encoding Categorical Variables

```python
# Encoding the target variable
label_encoder = LabelEncoder()
data['target'] = label_encoder.fit_transform(data['target'])
```

# 5. Feature Engineering
## 5.1 Creating New Features

```python
# Example: Creating a new feature from existing ones
data['sepal_area'] = data['sepal length (cm)'] * data['sepal width (cm)']
data['petal_area'] = data['petal length (cm)'] * data['petal width (cm)']
```

## 5.2 Transforming Features

```python
# Standardizing numerical features
scaler = StandardScaler()
numerical_cols = data.select_dtypes(include=['int64', 'float64']).columns
data[numerical_cols] = scaler.fit_transform(data[numerical_cols])
```

## 5.3 Dimensionality Reduction

```python
# Applying PCA
pca = PCA(n_components=3)
data_pca = pca.fit_transform(data[numerical_cols])

# Adding PCA components to the dataframe
data['pca1'] = data_pca[:, 0]
data['pca2'] = data_pca[:, 1]
data['pca3'] = data_pca[:, 2]
```

## 5.4 Feature Selection

```python
# SelectKBest
X = data.drop('target', axis=1)
y = data['target']

# Using chi2 for feature selection
select_k_best = SelectKBest(score_func=chi2, k=4)
X_new = select_k_best.fit_transform(X, y)
```

## 5.5 Recursive Feature Elimination (RFE)

```python
# Using RFE with a RandomForestClassifier
model = RandomForestClassifier()
rfe = RFE(model, 4)
fit = rfe.fit(X, y)

# Displaying the selected features
selected_features = X.columns[fit.support_]
selected_features
```

# 6. Model Building and Evaluation
## 6.1 Train-Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X[selected_features], y, test_size=0.3, random_state=42)
```

## 6.2 Model Training

```python
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

## 6.3 Model Evaluation

```python
# Predict on the test set
y_pred = model.predict(X_test)

# Calculate accuracy
accuracy = np.mean(y_pred == y_test)
print(f'Accuracy: {accuracy:.2f}')
```

# 7. 3D Plotting
## 7.1 3D Plot of Original Features

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data['sepal length (cm)'], data['sepal width (cm)'], data['petal length (cm)'], c=data['target'], cmap='viridis')
ax.set_xlabel('Sepal Length')
ax.set_ylabel('Sepal Width')
ax.set_zlabel('Petal Length')
plt.show()
```

## 7.2 3D Plot of PCA Components

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data['pca1'], data['pca2'], data['pca3'], c=data['target'], cmap='viridis')
ax.set_xlabel('PCA1')
ax.set_ylabel('PCA2')
ax.set_zlabel('PCA3')
plt.show()
```

# 8. Results

* Created sepal_area and petal_area as new features.
* Standardized all numerical features.
* Extracted three principal components using PCA.
* Selected top 4 features using SelectKBest.
* Identified top 4 features using RFE with RandomForest.
* Achieved 97% accuracy with RandomForestClassifier.
* 3D scatter plot of original features showed distinct species clusters.
* 3D scatter plot of PCA components revealed clear species separation.
* Key features like petal length and petal width were most important.
3D plots confirmed effective feature engineering for species differentiation.
